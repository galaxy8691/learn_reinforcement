```python
from PIL import Image
import numpy as np
def display_image(imageArray):
    display.display(Image.fromarray(np.array(imageArray),'RGB'),clear=True)
```

```python
from collections import namedtuple
Transition  = namedtuple('Transition', ('state','action', 'state_next','reward'))
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.fc1 = nn.Linear(4,512)
        self.fc2 = nn.Linear(512,512)
        self.fc3_adv = nn.Linear(512,2)
        self.fc3_val = nn.Linear(512,1)

    def forward(self,x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        adv = self.fc3_adv(x)
        val = self.fc3_val(x).expand(-1,2)
        x = val + adv - adv.mean(1,keepdim=True).expand(-1,2)
        return x
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MAX_STEP = 500
NUM_EPISODE = 500
CAPACITY = 10000
BATCH_SIZE = 32
GAMMA = 0.99
TD_ERROR_EPSILON = 0.0001
memory = []
td_memory = []
index = 0
target_count = 100
target_suceess_count = 0

main_model = Net()
main_model.to(device)
target_model = Net()
target_model.to(device)

optimizer = optim.Adam(main_model.parameters(), lr=0.0001)

def get_prioritized_indexs(batch_size):
    sum_absolute_td_error = np.sum(np.abs(td_memory))
    sum_absolute_td_error += TD_ERROR_EPSILON * len(td_memory)
    rand_list = np.random.uniform(0, sum_absolute_td_error, batch_size)
    rand_list = np.sort(rand_list)
    indexes = []
    idx = 0
    tmp_sum_absolute_td_error = 0
    for rand_num in rand_list:
        while tmp_sum_absolute_td_error < rand_num:
            tmp_sum_absolute_td_error += (abs(td_memory[idx]) + TD_ERROR_EPSILON)
            idx += 1
        if idx >= len(td_memory):
            idx = len(td_memory) - 1
        indexes.append(idx)
    return indexes

def push(state, action, state_next, reward):
    global index
    #print(index)
    if len(memory) < CAPACITY:
        memory.append(None)
    memory[index] = Transition(state,action,state_next, reward)
    index = (index + 1) % CAPACITY

def sample(batch_size):
    return random.sample(memory,batch_size)

def update_td_memory():
    global td_memory
    batch = Transition(*zip(*memory))
    state_batch = torch.cat(batch.state).to(device)
    action_batch = torch.cat(batch.action).to(device)
    reward_batch = torch.cat(batch.reward).to(device)
    non_final_next_states = torch.cat([s for s in batch.state_next if s is not None]).to(device)
    main_model.eval()
    target_model.eval()
    state_action_value = main_model(state_batch).gather(1,action_batch)
    mask = torch.ByteTensor(tuple(map(lambda s: s is not None, batch.state_next))).to(device) != 0
    nsv = torch.zeros(len(memory)).to(device)
    nsv[mask] = target_model(non_final_next_states).max(1)[0].detach().to(device)
    tds = reward_batch + GAMMA * nsv - state_action_value.squeeze()
    td_memory = tds.detach().numpy().tolist()

def replay():
    if len(memory) < BATCH_SIZE:
        return
    rand = random.random()
    if rand < 0.5:
        indexes = get_prioritized_indexs(BATCH_SIZE)
        transition = [memory[n] for n in indexes]
    else:
        transition = sample(BATCH_SIZE)
    batch = Transition(*zip(*transition))
    state_batch = torch.cat(batch.state).to(device)
    action_batch = torch.cat(batch.action).to(device)
    reward_batch = torch.cat(batch.reward).to(device)
    non_final_next_states = torch.cat([s for s in batch.state_next if s is not None]).to(device)
    main_model.eval()
    target_model.eval()
    state_action_value = main_model(state_batch).gather(1,action_batch)
    mask = torch.ByteTensor(tuple(map(lambda s: s is not None, batch.state_next))).to(device) != 0
    nsv = torch.zeros(BATCH_SIZE).to(device)
    nsv[mask] = target_model(non_final_next_states).max(1)[0].detach().to(device)
    expected_state_action_values = reward_batch + GAMMA * nsv
    main_model.train()
    loss = F.smooth_l1_loss(state_action_value, expected_state_action_values.unsqueeze(1))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

def update_target_model():
    target_model.load_state_dict(main_model.state_dict())   
    
def reset(env):
      observation = env.reset()[0]
      return observation

def reset_memory():
    global memory
    global index
    memory = []
    index = 0

def get_action(s,epsilon):
    if np.random.rand() < epsilon:
        action = torch.LongTensor([[random.randrange(2)]]).to(device)
    else:
        main_model.eval()
        with torch.no_grad():
            action = main_model(s).max(1)[1].view(1,1)
    return action

# def Q_learning(s, a, r, s_next, Q, eta, gamma, is_terminated):
#     s = digitize_state(s)
#     s_next = digitize_state(s_next)
#     if is_terminated:
#         Q[s, a] = Q[s, a] + eta * (r- Q[s, a])
#     else:
#         Q[s, a] = Q[s, a] + eta * (r + gamma * np.nanmax(Q[s_next, :]) - Q[s, a])
#     return Q

def run_once(env, initial_observation, epsilon, show=False, train=True):
    global target_count, target_suceess_count
    observation = initial_observation
    s = torch.from_numpy(observation).type(torch.FloatTensor)
    s = torch.unsqueeze(s,0).to(device)
    a = get_action(s, epsilon)
    a_next = a
    count = 0;
  
    while True:
        is_terminated = False
        if show:
            time.sleep(0.01)
            img = env.render()
            display_image(img)
        a = a_next
        #s_a_history[-1][1] = a
        s_next, reward, terminated, truncated, info = env.step(a.item())
        s_next = torch.from_numpy(s_next).type(torch.FloatTensor)
        s_next = torch.unsqueeze(s_next,0).to(device)
        #s_a_history.append([s_next, np.nan])
        if count > target_count:
            r = torch.FloatTensor([1.0])
            a_next = None
            is_terminated = True
            s_next = None
            target_suceess_count += 1
            if target_count <= 900:
                if target_suceess_count > 10:
                    target_count += 100
                    target_suceess_count = 0
                    print("new target is ", target_count)
        elif count < target_count and terminated:
            r = torch.FloatTensor([-1.0])
            a_next = None
            is_terminated = True
            s_next = None
            target_suceess_count = 0
        else:
            r = torch.FloatTensor([0.0])
            a_next = get_action(s_next, epsilon)
        push(s, a, s_next, r)
        if train:
            
            replay()
            
        if is_terminated:
            if show:
                time.sleep(0.01)
                img = env.render()
                display_image(img)
            if train:
                update_target_model()
                update_td_memory()
            break
        else:
            s = s_next
        count += 1
        
```
